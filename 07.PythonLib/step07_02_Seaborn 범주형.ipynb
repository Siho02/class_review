```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()

# Load the example flights dataset and convert to long-form
flights_long = sns.load_dataset("flights")
```

```python
flights_long
```

```python
flights_long['year'].value_counts()
```

```python
flights = flights_long.pivot("month", "year", "passengers")
flights
```

```python
# Draw a heatmap with the numeric values in each cell
# 중요 
# 섬세한 작업은 sns가 직접적으로 적용 불가한 부분 있음
# >> 해결책 : matplotlib

# https://seaborn.pydata.org/generated/seaborn.heatmap.html?highlight=heatmap#seaborn.heatmap

f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax);
```

```python
# annot = True : 각 데이터 표현, 생략시 수치값이 안보임

f, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(flights, fmt="d", linewidths=.5, ax=ax);
```

```python
# f, ax 생략시 최소 사이즈로 headmap이 그려짐, >> 사이즈 조절 

sns.heatmap(flights, annot = True, fmt="d", linewidths=.5);  
```

```python
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot = True, fmt="d");  
```

```python
# fmt = "d" >> 정수 표기로 출력하라는 포맷 설정
f, ax = plt.subplots(figsize=(15, 6))
sns.heatmap(flights, annot = True);  
```
