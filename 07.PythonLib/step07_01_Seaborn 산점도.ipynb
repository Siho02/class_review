데이터 참고 사이트


변수(피처)들 의미

1. total_bill - 팁을 준 사람들이 레스토랑에서 계산한 금액
2. tip - 팁 급액
3. sex - 성별
4. smoker - 흡연 여부
5. day - 요일
6. time - 식사 시간(팁을 받은 시간)
7. size - 식사한 인원수

```python
# 레스토랑 팁 데이터 셋 활용
tips = sns.load_dataset("tips")
tips
```

```python
tips.columns
```

```python
# total_bill, 즉 계산한 
# sns.relplot(x='total_bill', y='tip', data=tips) >> 밑에 <seaborn.axisgrid.FacetGrid at 0x280c19e1f0> 출력
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용
sns.set_style(style = 'darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips);
```

```python
# 스타일 적용
# feature를 색으로 구분 >> 흡연 비흡연에 대한 색을 구분 -- 범례 표현 설정도 가능
sns.set_style(style = 'darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'smoker');
```

```python
sns.set_style(style = 'darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'size');
```

```python
sns.set_style(style = 'darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'size', style = 'smoker');
```

```python
sns.set_style(style = 'darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'size', style = 'smoker', palette = 'Blues');
```

```python
sns.set_style(style = 'darkgrid')
sns.relplot(x='total_bill', y='tip', data=tips, hue = 'size', style = 'smoker', size = 'size');
```
