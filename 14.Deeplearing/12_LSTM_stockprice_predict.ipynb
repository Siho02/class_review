### X, y 

```python
df_y = df['Close'].to_frame() #Series.to_frame() >> Series(1) Dataframe(2) 
df_X = df
```

### Scaling - MinMaxScaler
- Scaler X, y   .
    - y inverse  

```python
from sklearn.preprocessing import MinMaxScaler
scaler_X = MinMaxScaler() #df_X 
scaler_y = MinMaxScaler() #df_y  

X = scaler_X.fit_transform(df_X)
y = scaler_y.fit_transform(df_y)
```

##    
- X: 50 , y: 51 . 
    - 50   51  .
     

X, y .
 index window size   .

```python
X.shape, y.shape, type(X)
```

```python
window_size = 50 #50       
data_X = []      #input(feature)   
data_y = []      #output(Label - target)   

for i in range(0, y.size - window_size): #y.size - window_size  index 50     x
    _X = X[i:i+window_size]
    _y = y[i+window_size]
    data_X.append(_X)
    data_y.append(_y)
    
```

## Train, Test 

```python
train_size = int(len(data_X) * 0.8)
test_size = len(data_X) - train_size
train_size, test_size, train_size+test_size
```

```python
X_train = np.array(data_X[:train_size])
X_test = np.array(data_X[train_size:])

y_train = np.array(data_y[:train_size])
y_test = np.array(data_y[train_size:])

X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

####   

```python
LEARNING_RATE = 0.001
N_EPOCHS = 100
N_BATCHS = 100

N_TRAIN = X_train.shape[0]
N_TEST = X_test.shape[0]
```

#### dataset

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

import numpy as np

np.random.seed(0)
tf.random.set_seed(0)
```

```python
train_dataset = tf.data.Dataset.from_tensor_slices((X_train, y_train)).shuffle(N_TRAIN).batch(N_BATCHS, drop_remainder=True)
test_dataset = tf.data.Dataset.from_tensor_slices((X_test, y_test)).batch(N_BATCHS)
```

## Model 
- LSTM  return_sequences               . 
    - return_sequences=True:   . False():  Sequence    
- many to many   LSTM     return_sequence=True  . 
- Dese False   .
-    return_sequences=False ,  return_sequence=True  .



<center>[: https://tykimos.github.io/2017/04/09/RNN_Getting_Started]</center>

```python
def create_model():
    model = keras.Sequential()
    
    model.add(layers.LSTM(32, activation='relu', return_sequences=False, input_shape=(50,5))) 
        # return_sequences=False :  dense 
        # return_sequences=True :   RNN(LSTM) 
    model.add(layers.Dense(32, activation='relu'))
    model.add(layers.Dense(1)) #
    
    return model
```

```python
model = create_model()
model.compile(optimizer = keras.optimizers.Adam(LEARNING_RATE), loss='mse')
model.summary()
```

```python
np.__version__
```

##  , 

```python
hist = model.fit(train_dataset, epochs = N_EPOCHS, validation_data = test_dataset)
```

```python
plt.plot(range(1,N_EPOCHS+1), hist.history['val_loss'])
plt.ylim(0, 0.05)
plt.show()
```

```python
model.evaluate(test_dataset)
```
